```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("cars.csv")
```

```python
x=np.arange(20)
y=np.sqrt(x)
```

```python
sns.lineplot(x,y)
plt.show()
```

```python
df.head()
```

```python
sns.distplot(df["width"])
plt.show()
```

```python
sns.countplot(data=df, x="body-style")
plt.show()
```

```python
sns.countplot(data=df, y="make")
plt.show()
```

```python
sns.countplot(data=df, x="body-style", hue = "fuel-type")
plt.show()
```

```python
sns.countplot(data=df, x="drive-wheels", hue = "fuel-type")
plt.show()
```

# Heatmap


```python
corr = df.corr()
```

```python
sns.heatmap(corr, annot = True)
plt.show()
```

```python
sns.heatmap(corr, annot = True, cmap = "viridis")
plt.show()
```

# ScatterPlot

```python
sns.scatterplot(data=df,x="width",y="price")
plt.plot()
```

```python
sns.scatterplot(data=df,x="width",y="price",hue="body-style")
plt.plot()
```

# BoxPlot

```python
sns.boxplot(data=df, x="price")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df, x="price",y="make")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df, x="price",y="body-style")
plt.show()
```

# Sub Plot

```python
fig, axes = plt.subplots(2,2)
```

```python
fig, axes = plt.subplots(2,2, figsize= (7,7))
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,10))
sns.distplot(df["price"], ax=axes[0,0])
sns.countplot(data=df, y="make",ax=axes[0,1])
sns.scatterplot(data=df,x="width",y="price",hue="body-style",ax=axes[1,0])
sns.boxplot(data=df, x="price",ax=axes[1,1])
plt.show()
```
